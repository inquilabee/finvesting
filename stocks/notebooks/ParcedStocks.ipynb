### Important Columns for Analysis

#### Auto Filtering

```python
MISSING_COLUMNS_THRESHOLD = 0.7

missing_percentage = df_parsed_stocks.isna().sum() / df_parsed_stocks.shape[0]
missing_cols = missing_percentage[missing_percentage > MISSING_COLUMNS_THRESHOLD].index

missing_cols.tolist()
```

#### Manual Filtering

```python
columns_to_delete = [
    "address1",
    "address2",
    "city",
    "zip",
    "country",
    "phone",
    "fax",
    "website",
    "company_officers",
    "financial_currency",
    "underlying_symbol",
    "short_name",
    "long_name",
    "exchange",
    "quote_type",
    "industry",
    "time_zone_full_name",
    "time_zone_short_name",
    "industry_disp",
    "sector",
    "sector_disp",
    "uuid",
    "message_board_id",
]
```

```python
columns_to_delete = set(columns_to_delete) | set(missing_cols)

df_parsed_stocks.drop(columns=columns_to_delete, inplace=True)
```

#### Deleted Columns

```python
print(list(columns_to_delete))
```

#### Available columns to select

```python
available_columns = df_parsed_stocks.columns.tolist()
```

```python
accepted_columns = sum([cols for cols in reorganized_columns.values()], [])

print(len(accepted_columns))
print(accepted_columns)
```

### Ignored Columns

```python
ignored_columns = set(available_columns) - set(accepted_columns)

print(len(ignored_columns))
print(list(ignored_columns))
```

### Saving Data

```python
df_parsed_stocks = df_parsed_stocks[
    sum([cols for cols in reorganized_columns.values()], [])
]
```
