```python
attributes = []

for attr in dir(ticker):
    if not attr.startswith(("_", "get")):
        try:
            attribute = getattr(ticker, attr)

            if callable(attribute):
                attribute()
            
            attributes.append(attr)
        except Exception as e:
            pass

attributes
```
